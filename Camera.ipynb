```python
import cv2
import numpy as np
import io
import time
import winsound as sd

def beepsound():
    fr = 2000
    du = 1000
    sd.Beep(fr, du)


def VideoWrite():
    try:
        print("카메라 구동")
        cap = cv2.VideoCapture(1)
    except:
        print("카메라 구동실패")
        return

    cap.set(cv2.CAP_PROP_AUTOFOCUS, 1)
    width = int(cap.get(3))
    height = int(cap.get(4))
    fourcc = cv2.VideoWriter_fourcc(*'DIVX')
    i = 0;

    # 비디오 저장을 위한 객체를 생성
    out = cv2.VideoWriter('C:\\Users\\HwaJong\\Desktop\\SaveVideo.mp4',fourcc,20.0,(width, height))

    endTime = time.time() + 7
    while(time.time() < endTime):
        ret, frame = cap.read()
        
        if i == 0: 
            beepsound()
        i += 1

        if not ret:
            print("비디오 읽기 오류")
            break

        # 비디오 프레임이 정확하게 촬영되었으면 화면에 출력
        cv2.imshow('video',frame)
        # 비디오 프레임이 제대로 출력되면 해당파일에 프레임을 저장
        out.write(frame)

    # 비디와 관련 장치들을 다 닫아줌.
    print('녹화 종료')
    beepsound()
    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

```python
VideoWrite()  # 비디오 촬영
```

```python
def allFrame():
    cap = cv2.VideoCapture('C:\\Users\\HwaJong\\Desktop\\SaveVideo.mp4') #동영상 불러오기
    fps = cap.get(cv2.CAP_PROP_FPS)
    # print(fps)

    if not cap.isOpened():
        print("Could not Open :")
        exit(0)

    num = 0 
    while(cap.isOpened()):
        ret, frame = cap.read() 
        if ret == False:
            break
        if ret:
            path = 'C:\\Users\\HwaJong\\Desktop\\frame\\' + str(num) + '.jpg' 
            cv2.imwrite(path, frame) #영상 -> 이미지로 저장 
        num += 1

    print('분할 종료')
    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2

def secFrame():
    filepath = "C:\\Users\\HwaJong\\Desktop\\videoTest\\SaveVideo.mp4"
    filepath2 ="C:\\Users\\HwaJong\\Desktop\\videoTest\\frame\\" 
    cap = cv2.VideoCapture(filepath)
    count = 0

    if not cap.isOpened():
        print("Could not Open :", filepath)
        exit(0)
        
    fps = cap.get(cv2.CAP_PROP_FPS)
    print("fps: ", fps)

    while(cap.isOpened()):
        ret, image = cap.read()
        if ret == False:
            break
        if(int(cap.get(1)) % fps == 0): #앞서 불러온 fps 값을 사용하여 1초마다 추출
            cv2.imwrite(filepath2 + "frame%d.jpg" % count, image)
            print('Saved frame number :', str(int(cap.get(1))))
            count += 1

    cap.release()
```
